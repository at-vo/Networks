```python
import numpy as np
import matplotlib.pyplot as plt
import math
from scipy import signal

img = plt.imread("image1.jfif")
```

```python
def grayScale(rgb):
    return np.dot(rgb[...,:3], [0.2989, 0.5870, 0.1140])

image1 = grayScale(img)
plt.imshow(image1, cmap = 'gray')
image1.shape
```

```python
#function for 2D gaussian value
def gaus(x, y, sigma):
    return np.exp(-1.0 * ((pow(x, 2) + pow(y, 2))/(2 * pow(sigma, 2)))) / (2 * math.pi * pow(sigma, 2))

# make a 2D Gaussian Kernel 
def makeKernel(sigma, size):
    size = int(size) // 2
    x, y = np.mgrid[-size:size+1, -size:size+1]
    kernel = gaus(x, y, sigma)
    kernel *= 1.0/np.sum(kernel)
    return kernel

# seperate the kernel into two 1D kernels
def seperate(kernel): 
    I,J = np.unravel_index(np.abs(kernel).argmax(), kernel.shape)
    f1 = kernel[:, J, None]
    f2 = kernel[None, I, :] / kernel[I, J]
    return f1, f2

# implement seperable convolve 
def sepConvolve(img, row, col):
    return signal.convolve2d(signal.convolve2d(img, row), col)

sigma = 1
size = 5

kernel = makeKernel(sigma, size)
rowFilter, colFilter = seperate(kernel)

# implement gaussian convolution as seperable sequence
output = sepConvolve(image1, rowFilter, colFilter)

plt.imshow(output, cmap = 'gray')
```

```python
def sobel(img):
    # sobel arrays
    sobX = np.array([[-1, 0, 1], [-2, 0, 2], [-1, 0, 1]])
    sobY = sobX.T
    
    # seperate them
    xRow, xCol = seperate(sobX)
    yRow, yCol = seperate(sobY)
    
    # compute x and y derivatives of image
    Ix = sepConvolve(img, xRow, xCol)
    Iy = sepConvolve(img, yRow, yCol)
        
    return Ix, Iy

Ix, Iy = sobel(output)

# compute magnitude of gradient at every pixel
G = np.hypot(Ix, Iy)

# normalize
G = G / G.max()

plt.imshow(G, cmap = 'gray')
plt.title("Normalized")
plt.xticks([])
plt.yticks([])
```

```python
def threshold(img, low, high):
    Ix, Iy = sobel(img)
    
    # adjust high and low threshold to image
    high *= img.max()
    low *= high
    
    # calculate gradient magnitude and direction
    gMag = np.hypot(Ix, Iy)    
    gDir = np.arctan2(Ix, Iy)
    
    zero = np.zeros(img.shape)
    
    strI, strJ = np.where(img >= high)
    zeroI, zeroJ = np.where(img < low)
    weakI, weakJ = np.where(((img <= high) & (img >=low)))
    
    zero[strI, strJ] = 255.
    zero[zeroI,zeroJ] = 0.
    
    return zero
    
low = 0.01
high = 0.1

thres = threshold(output, low, high)
plt.imshow(thres, cmap = 'gray')
```

```python
def hysteresis(img, weak, strong=255):
    M, N = img.shape  
    for i in range(1, M-1):
        for j in range(1, N-1):
            if (img[i,j] == weak):
                try:
                    if ((img[i+1, j-1] == strong) or (img[i+1, j] == strong) or (img[i+1, j+1] == strong)
                        or (img[i, j-1] == strong) or (img[i, j+1] == strong)
                        or (img[i-1, j-1] == strong) or (img[i-1, j] == strong) or (img[i-1, j+1] == strong)):
                        img[i, j] = strong
                    else:
                        img[i, j] = 0
                except IndexError as e:
                    pass
    return img

edges = hysteresis(output, low, high)
plt.imshow(edges, cmap = 'gray')
```
